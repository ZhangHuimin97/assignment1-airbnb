```python

stylewhole=[]
layout=[]
style=[]
name=[]
price=[]
policy=[]
comment=[]
equipment=[]


for i in soup.find_all('div',attrs={'class':"_8ssblpx"}):
    styleall=i.find('small', attrs={'class':"_f7heglr"}).text
    stylem=styleall.split()
    if len(stylem)<3:
        stylem=stylem.append(''+'')
        style2='none'
        layout1='none'
    else:   
        style2=stylem[0]
        layout1=stylem[2]
    
    stylewhole.append(stylem)
    layout.append(layout1)
    style.append(style2)
    nameall=i.find('div', attrs={'class':"_e296pg"}).text
    name.append(nameall)
    priceall=i.find('span', attrs={'class':"_xbdrhpb"}).text
    pricem=priceall[3:6]
    price.append(pricem)
    commentsall=i.find('span', attrs={'class':"_1cy09umr"}).text
    comment.append(commentsall)
    policyall=i.find('small', attrs={'class':"_mbm8v7"}).text
    policy.append(policyall)
    equipmentall=i.find('div',attrs={'class':"_2q1l5fu"}).text
    equipmentm=equipmentall.replace('等多项便利设施','')
    equipmentt=equipmentm.replace('·','')
    equipment.append(equipmentt)
    
    
```
